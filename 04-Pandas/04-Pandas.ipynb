**Pandas** is a powerful Python library for data manipulation and analysis. It is built on top of NumPy and provides two key data structures:

- Handle missing values
- Compute statistics and aggregations
- Group and pivot data

Let's start by importing the libraries we will need throughout this notebook.

````python
import pandas as pd
import numpy as np

print('pandas version:', pd.__version__)
print('numpy  version:', np.__version__)
````

## 1. Pandas Series

A **Series** is a one-dimensional sequence of values, each associated with an **index** label.  
Think of it as a Python list plus a custom index.

```
index:   1       2       3
values:  0.3     0.5     0.8

index:  '3-July' '4-July' '5-July'
values:  0.3      0.5      0.8
```

### 1.1 Creating a Series

````python
# From a list — index is set automatically (0, 1, 2, ...)
s1 = pd.Series([2.0, 3.1, 4.5])
print(s1)
````

````python
# From a list with an explicit index
s2 = pd.Series([2.0, 3.1, 4.5], index=['mon', 'tue', 'wed'])
print(s2)
````

````python
# From a dictionary — keys become the index
s3 = pd.Series({'c': 2.0, 'b': 3.1, 'a': 4.5})
print(s3)
````

### 1.2 Accessing values and index

Every Series exposes `.values` (a NumPy array) and `.index`.

````python
s = pd.Series([2.0, 3.1, 4.5], index=['mon', 'tue', 'wed'])
print('values:', s.values)   # NumPy array
print('index: ', s.index)
````

### 1.3 Indexing a Series

Pandas offers two indexers:

| Indexer | Based on | Slice stop |
|---------|----------|------------|
| `.loc`  | **explicit** labels | **included** |
| `.iloc` | **implicit** position (0-based) | **excluded** |

````python
s = pd.Series([2.0, 3.1, 4.5], index=['a', 'b', 'c'])

print(s.loc['a'])    # explicit index
print(s.iloc[0])     # implicit index (position)

# You can also assign values
s.loc['b'] = 10
print(s)
````

````python
# Slicing — note the different stop-inclusion rules!
s = pd.Series([2.0, 3.1, 4.5], index=['a', 'b', 'c'])

print(s.loc['a':'c'])   # stop label 'c' IS included
print()
print(s.iloc[1:3])      # stop position 3 is NOT included
````

````python
# Boolean masking — select elements that satisfy a condition
s = pd.Series([2.0, 3.1, 4.5], index=['a', 'b', 'c'])
print(s[(s > 2) & (s < 10)])
````

````python
# Fancy indexing — pass a list of labels or positions
s = pd.Series([2.0, 3.1, 4.5], index=['a', 'b', 'c'])
print(s.loc[['a', 'c']])
print()
print(s.iloc[[0, 2]])
````

## 2. Pandas DataFrame

A **DataFrame** is a 2-dimensional table where each column is a Series sharing the same index.

```
         Price   Quantity   Liters
Water      1.0          5      1.5
Beer       1.4         10      0.3
Wine       5.0          8      1.0
```

### 2.1 Creating a DataFrame

````python
# From a dictionary of Series
price    = pd.Series([1.0, 1.4, 5.0], index=['a', 'b', 'c'])
quantity = pd.Series([5,   10,  8  ], index=['a', 'b', 'c'])
liters   = pd.Series([1.5, 0.3, 1.0], index=['a', 'b', 'c'])

df = pd.DataFrame({'Price': price, 'Quantity': quantity, 'Liters': liters})
print(df)
````

````python
# From a dictionary of lists — index is automatic
dct = {'c1': [0, 1, 2], 'c2': [0, 2, 4]}
df2 = pd.DataFrame(dct)
print(df2)
````

````python
# From a list of dictionaries — each dict is one row
dict_list = [{'c1': i, 'c2': 2*i} for i in range(3)]
df3 = pd.DataFrame(dict_list)
print(df3)
````

````python
# From a 2D NumPy array
arr = np.arange(6).reshape((3, 2))
df4 = pd.DataFrame(arr, columns=['c1', 'c2'], index=['a', 'b', 'c'])
print(df4)
````

### 2.2 Inspecting a DataFrame

````python
df = pd.DataFrame({'Price':    [1.0, 1.4, 5.0],
                   'Quantity': [5,   10,  8  ],
                   'Liters':   [1.5, 0.3, 1.0]},
                  index=['a', 'b', 'c'])

print('Columns:', df.columns.tolist())
print('Index:  ', df.index.tolist())
print('Values (NumPy array):\n', df.values)
````

### 2.3 Accessing a DataFrame

The same `.loc` / `.iloc` indexers work for DataFrames.  
The syntax is `df.loc[row_selector, column_selector]`.

````python
# Access a single column — returns a Series
print(df['Quantity'])
````

````python
# Access a single row by explicit / implicit index
print(df.loc['a'])    # explicit
print()
print(df.iloc[0])     # implicit
````

````python
# Slicing — select a range of rows AND columns
print(df.loc['b':'c', 'Quantity':'Liters'])
````

````python
# Boolean masking — select rows based on a condition
mask = (df['Quantity'] < 10) & (df['Liters'] > 1)
print(df.loc[mask, 'Quantity':])   # masking + slicing
````

````python
# Fancy indexing — pick arbitrary rows and columns by name
print(df.loc[['a', 'c'], ['Price', 'Liters']])
````

### 2.4 Modifying a DataFrame

````python
df_mod = df.copy()

# Assign a value to a selection
df_mod.loc[['a', 'c'], ['Price', 'Liters']] = 0
print(df_mod)
````

````python
# Add a new column (from a list or a Series)
df_mod = df.copy()
df_mod['Available'] = [True, False, True]
print(df_mod)
````

````python
# Drop columns - returns a copy by default
df_dropped = df_mod.drop(columns=['Quantity', 'Liters'])
print(df_dropped)
````

````python
# Rename columns using a mapping dictionary
df_renamed = df_mod.rename(columns={'Quantity': 'nItems', 'Liters': '[L]'})
print(df_renamed)
````

## 3. I/O — Reading and Writing Data

Pandas can read and write many file formats. The most common is **CSV**.

### 3.1 Reading a CSV file

````python
# Create a small sample CSV to work with
csv_content = """c1,c2,c3
0,1,2

3,4,5
6,7,8
"""
with open('sample.csv', 'w') as f:
    f.write(csv_content)
````

````python

df_na = pd.read_csv('sample_missing.csv', na_values=['no info', 'x'])
print(df_na)
print('\nData types:\n', df_na.dtypes)
````

### 3.2 Writing a CSV file

````python
# Save to CSV — use index=False to avoid writing the row index
df_csv.to_csv('output.csv', sep=',', index=False)

# Verify by reading it back
print(pd.read_csv('output.csv'))
````

## 4. Computation with Pandas

### 4.1 Unary operations

NumPy universal functions (`ufuncs`) work directly on Series and DataFrames, element-wise.

````python
my_series = pd.Series([1.0, 2.0, 3.0, 4.0])

print(my_series / 4 + 1)
print()
print(np.exp(my_series))
````

### 4.2 Operations between Series

Arithmetic operations align on the **index** automatically.  
Elements whose index does not match in both Series will produce **NaN**.

````python
s1 = pd.Series([1, 3, 10], index=['a', 'b', 'c'])
s2 = pd.Series([1, 3, 30], index=['a', 'b', 'd'])

print(s1 + s2)  # 'c' and 'd' have no match → NaN
````

### 4.3 Operations between DataFrames

The same alignment logic applies to both **index** and **columns**.

````python
df1 = pd.DataFrame({'Total': [3, 1, 10], 'Quantity': [4, 2, 20]},
                   index=['b', 'a', 'c'])
df2 = pd.DataFrame({'Total': [1, 3, 30], 'Quantity': [2, 4, 40]},
                   index=['a', 'b', 'd'])

print(df1 + df2)  # row 'c' and 'd' have no match → NaN
````

### 4.4 Aggregation functions

For DataFrames, aggregation functions are applied **column-wise** and return a Series.

````python
data = pd.DataFrame({'Total':    [1, 3, 5],
                     'Quantity': [2, 4, 6]},
                    index=['a', 'b', 'c'])

print('mean:\n', data.mean())
print('\nstd:\n', data.std())
print('\nmin:\n', data.min())
````

````python
# Practical example: z-score normalization
mean_s = data.mean()
std_s  = data.std()
data_norm = (data - mean_s) / std_s
print(data_norm)
````

## 5. Missing Values

Pandas represents missing data using:
- `None` — Python's null object (any dtype)
- `np.nan` — a floating-point sentinel (preferred for numerical data, better performance)

Pandas automatically converts between them when needed.

````python
s = pd.Series([4, None, 5, np.nan])
print(s)          # Both None and np.nan become NaN
print('dtype:', s.dtype)
````

````python
# isnull() — boolean mask of missing values
print(s.isnull())
print()
# notnull() — the inverse
print(s.notnull())
````

````python
# dropna() — remove missing values
print(s.dropna())
````

````python
# dropna() on a DataFrame: by default removes any row with at least one NaN
df_miss = pd.DataFrame({'Total':    [1, 3,   5],
                        'Quantity': [2, None, 6]},
                       index=['a', 'b', 'c'])
print('Original:\n', df_miss)
print('\nAfter dropna():\n', df_miss.dropna())
````

````python
# fillna() — replace NaN with a fixed value
print(s.fillna(0))
````

````python
# fillna() with forward fill (ffill) — propagate the last valid value
print(s.ffill())
````

## 6. Grouping Data

Pandas `groupby` is the equivalent of SQL `GROUP BY`.  
The typical workflow is: **split** → **apply** → **combine**.

1. Split the DataFrame into groups based on a key column
2. Apply a function to each group (aggregation, filter, transform)
3. Combine the results

````python
df_g = pd.DataFrame({'k':  ['a', 'b', 'a', 'b'],
                     'c1': [2,   10,  3,   15],
                     'c2': [4,   20,  5,   30]})
print(df_g)

grouped = df_g.groupby('k')   # creates a DataFrameGroupBy object
````

````python
# Iterate over groups
for key, group_df in grouped:
    print(f'--- Group: {key} ---')
    print(group_df)
````

````python
# Aggregate all columns — result index is the group key
print(grouped.mean())
````

````python
# Aggregate a single column — returns a Series
print(grouped['c1'].mean())
````

````python
# Filter groups: keep only groups where mean of c1 > 5
result = grouped.filter(lambda x: x['c1'].mean() > 5)
print(result)
````

## 7. Combining DataFrames

### 7.1 Concatenation

`pd.concat()` stacks Series or DataFrames along an axis.  
Indices are preserved (duplicates are allowed).

````python
s1 = pd.Series(['a', 'b'], index=[1, 2])
s2 = pd.Series(['c', 'd'], index=[1, 2])

# Default: keep original indices (may create duplicates)
print(pd.concat((s1, s2)))
````

````python
# Use ignore_index=True to get a fresh numeric index
print(pd.concat((s1, s2), ignore_index=True))
````

````python
# Concatenating DataFrames — vertical by default
df1 = pd.DataFrame({'Total': [1, 3], 'Quantity': [2, 4]}, index=['a', 'b'])
df2 = pd.DataFrame({'Total': [5, 7], 'Quantity': [6, 8]}, index=['c', 'd'])

print(pd.concat((df1, df2)))
````

````python
# If columns differ, missing values are filled with NaN
df3 = pd.DataFrame({'Total': [5, 7], 'Quantity': [6, 8], 'Liters': [1.0, 2.0]},
                   index=['c', 'd'])
print(pd.concat((df1, df3)))
````

### 7.2 Merging (Relational Join)

`pd.merge()` joins two DataFrames like a SQL JOIN.  
By default it merges on columns with the **same name** in both DataFrames.

````python
# One-to-one merge
left  = pd.DataFrame({'k1': [0, 1], 'c2': ['a', 'b']}, index=['i1', 'i2'])
right = pd.DataFrame({'k1': [1, 0], 'c3': ['b1', 'a1']}, index=['i1', 'i2'])

print(pd.merge(left, right))   # merge on 'k1'
````

````python
# Many-to-one merge
left_m = pd.DataFrame({'k1': [0, 1, 0, 1], 'c2': ['a', 'b', 'c', 'd']})
right_m = pd.DataFrame({'k1': [1, 0], 'c3': ['b1', 'a1']})

print(pd.merge(left_m, right_m))
````

## 8. Pivoting

`pivot_table()` is useful to summarise a dataset by two categorical variables — very similar to Excel pivot tables.

````python
# Sensor failure dataset
df_piv = pd.DataFrame({
    'type':  ['a','b','b','a','b','a','b','a'],
    'class': [3,  2,  3,  3,  2,  1,  1,  2],
    'fail':  [1,  1,  1,  0,  1,  0,  0,  0]
})
print(df_piv)
````

````python
# Total number of failures per (type, class)
df_piv.pivot_table('fail', index='type', columns='class', aggfunc='sum')
````

````python
# Failure rate per (type, class)
df_piv.pivot_table('fail', index='type', columns='class', aggfunc='mean')
````

## 9. Multi-Index

A **Multi-Index** adds a hierarchy of labels to rows (or columns), letting you index data by multiple levels at once.  
A classic use case: data indexed by both *city* and *year*.

### 9.1 Multi-indexed Series

````python
ix = [['Rome', 'Rome', 'Turin', 'Turin'],
      ['2018', '2019', '2018', '2019']]

s_mi = pd.Series([10, 13, 7, 9], index=ix)
s_mi = s_mi.sort_index()          # always sort a multi-index before slicing!
s_mi.index.names = ['city', 'year']
print(s_mi)
````

````python
# Access by outer level
print(s_mi.loc['Turin'])
````

````python
# Access a specific inner level across all outer levels
print(s_mi.loc[:, '2018'])   # all cities, only year 2018
````

````python
# Boolean masking still works
print(s_mi[s_mi > 10])
````

### 9.2 Multi-indexed DataFrame

````python
ix   = [['Rome', 'Rome', 'Turin', 'Turin'], ['2018', '2019', '2018', '2019']]
cols = [['c1','c1','c2','c2'], ['a','b','a','b']]
data = np.arange(16).reshape((4, 4))

df_mi = pd.DataFrame(data, index=ix, columns=cols)
print(df_mi)
````

````python
# Access outer column level
print(df_mi['c1'])
````

````python
# Access both column levels
print(df_mi['c1', 'a'])
````

````python
# pd.IndexSlice for precise multi-index slicing
ix_sl = pd.IndexSlice
print(df_mi.loc[ix_sl['Rome', '2018'], ix_sl['c1':'c2', 'a']])
````

### 9.3 `reset_index` and `set_index`

````python
# reset_index — turns the multi-index into regular columns
df_mi.index.names = ['city', 'year']
df_reset = df_mi.reset_index()
print(df_reset)
````

````python
# set_index — inverse of reset_index
df_back = df_reset.set_index(['city', 'year'])
print(df_back)
````

### 9.4 `unstack` and `stack`

````python
# unstack — move an index level to columns (multi-index Series → DataFrame)
print(s_mi.unstack())
````

````python
# stack — inverse: move a column level to the index (DataFrame → multi-index Series)
print(s_mi.unstack().stack())
````
